1. Instruction (For Data Cleaning)

   • Find all null value in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data['Cylinders'].fillna(data['Cylinders'].mean(), inplace = True)
```

```python
numeric_columns = data.select_dtypes(include='number').columns
data[numeric_columns] = data[numeric_columns].fillna(data[numeric_columns].mean())
```

```python
non_numeric_columns = data.select_dtypes(exclude='number').columns
data[non_numeric_columns] = data[non_numeric_columns].fillna(data[non_numeric_columns].mode().iloc[0])
```
